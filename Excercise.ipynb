```python
df = df.drop(columns=['Name', 'Ticket', 'Embarked', 'PassengerId','Cabin'])
df.Age = df.Age.fillna(df.Age.mean()).astype(int)
df.Sex = df.Sex.replace({'female':1, 'male':0})
df.head(10)
```

```python
target = df.pop("Survived")
df.head(10)
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.3)
display(X_train.head())
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_proba = lr.predict_proba(X_test)
y_proba_rand = [x if x>y else y for x, y in y_proba ]
y_proba_2 = [x for x, y in y_proba ]

result = pd.DataFrame({'Prediction': y_pred, 'Actual':y_test, "Predict_Proba":y_proba_rand, "proba 0": y_proba_2})

result.head()
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
def plot_confusion_matrix(cm, target_names, title = 'The Confusion Matrix', cmap = plt.cm.YlOrRd):
    plt.imshow(cm, interpolation = 'nearest', cmap = cmap)
    plt.tight_layout()
    
    # Add feature labels to x and y axes
    tick_marks = np.arange(len(target_names))
    plt.xticks(tick_marks, target_names, rotation=45)
    plt.yticks(tick_marks, target_names)
    
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    
    plt.colorbar()
    
plot_confusion_matrix(cm, ["Female", "Male"])
```

```python
from sklearn import model_selection


lr = LogisticRegression()
# lr.fit(X_train, y_train)

score = model_selection.cross_val_score(lr, X_train, y_train)
score
```

```python
from sklearn.metrics import classification_report


print(classification_report(result.Actual, result.Prediction))
```

```python
fare = df.pop('Fare')
age = df.pop('Age')
pclass = df.pop('Pclass')
sex = df.pop('Sex')
```

```python
age_inter = [0, 12, 18, 25, 50, 100]
age_labels = ['Kid', 'Teen','Young_Adult', 'Adult', 'Elderly']
age_dum = pd.get_dummies(pd.cut(age, age_inter, labels=age_labels, include_lowest=True))

pclass_labels = {1:'Upper', 2:'Middle', 3:'Lower'}
pclass = pclass.replace(pclass_labels)
pclass_dum = pd.get_dummies(pclass)

sex_labels = {1:'Female', 0:'Male'}
sex = sex.replace(sex_labels)
sex_dum = pd.get_dummies(sex)


df = pd.concat([df, age_dum, pclass_dum, sex_dum], axis=1)

df.sample(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.3, random_state=12)
# display(X_train.head())


lr2 = LogisticRegression(penalty='l1',tol=0.0001, C=0.9, random_state=12)
lr2.fit(X_train, y_train)

y_pred = lr2.predict(X_test)
y_proba = lr2.predict_proba(X_test)
y_proba_rand = [x if x>y else y for x, y in y_proba ]
y_proba_2 = [x for x, y in y_proba ]

result2 = pd.DataFrame({'Prediction': y_pred, 'Actual':y_test, "Predict_Proba":y_proba_rand, "proba 0": y_proba_2})

result2.head()

```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
plot_confusion_matrix(cm, ["Survived", "X.X"])
```

```python
score = model_selection.cross_val_score(lr2, X_train, y_train)
score
```
